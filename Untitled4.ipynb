```python
import os.path

import numpy as np
import cv2
import sys
import matplotlib.pyplot as plt

IMAGE_PATH = "lenna.jpg"
assert os.path.exists(IMAGE_PATH)
image = cv2.imread(IMAGE_PATH)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
histogram = cv2.calcHist([gray_image], [0], None, [256], [0, 256])
histogram = np.squeeze(histogram)
histogram_x_range = list(range(len(histogram)))

fig, ax = plt.subplots(1, 2)
ax[0].imshow(gray_image, "gray")
ax[0].set_title("Image")
ax[1].bar(x=histogram_x_range, height=histogram)
ax[1].set_title("Histogram")
```

```python
import numpy as np
import sys
import cv2
import matplotlib.pyplot as plt

FULL_PATH = "train_crop.jpg"
TARGET_PATH = "full_image.jpg"
assert os.path.exists(TARGET_PATH)
assert os.path.exists(FULL_PATH)

def images_in_rgb(file:str) -> np.ndarray:
  image = cv2.imread(file)
  rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  return rgb_image

def image_in_hsv(file: str) -> np.ndarray:
  image = cv2.imread(file)
  hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
  return hsv_image

train_hsv = image_in_hsv(FULL_PATH)
train_hsv_h, train_hsv_w = train_hsv.shape[:2]
train_hsv_area = train_hsv_h * train_hsv_h

full_hsv = image_in_hsv(TARGET_PATH)

train_im_hist = cv2.calcHist([train_hsv], [0, 1], None, [180, 256], [0, 180, 0, 256])
backprojection_map = cv2.calcBackProject(
[full_hsv], [0, 1], train_im_hist, [0, 180, 0, 256], scale=1
)
# Переведем эти значения в вероятности
backprojection_map = backprojection_map.astype(np.float32)
backprojection_map /= train_hsv_area
#plt.imshow(backprojection_map, cmap="inferno")

threshold = 0.00001
full_image_rgb = images_in_rgb(TARGET_PATH)
mask = backprojection_map > threshold
masked_image = np.copy(full_image_rgb)
masked_image[mask == 0] = [0, 0, 0]
fig, (image1_axes, image2_axes) = plt.subplots(1, 2, figsize=(100, 50))
image1_axes.imshow(full_image_rgb)
image2_axes.imshow(masked_image)
```

```python
from itertools import accumulate
import numpy as np
import os
import cv2
import matplotlib.pyplot as plt

PATH1SKIN = "1skin.jpg"
assert os.path.exists(PATH1SKIN)
PATH2SKIN = "2skin.jpg"
assert os.path.exists(PATH2SKIN)
PATHIMAGE = "train.jpg"
assert os.path.exists(PATHIMAGE)

def images_in_rgb_format(file:str) -> np.ndarray:
  image = cv2.imread(file)
  image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  return image

def images_in_hsv_format(file:str) -> np.ndarray:
  image = cv2.imread(file)
  image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
  return image

skin1 = images_in_hsv_format(PATH1SKIN)
skin1_h, skin1_w = skin1.shape[:2]
skin_area = skin1_h * skin1_w

skin2 = images_in_hsv_format(PATH1SKIN)
skin2_h, skin1_w = skin1.shape[:2]
skin_area = skin1_h * skin1_w

train_hsv = images_in_hsv_format(PATHIMAGE)

skin1_hist = cv2.calcHist([skin1], [0], None, [skin_area], [0, 180], accumulate=False)
skin2_hist = cv2.calcHist([skin2], [0], None, [skin_area], [0, 180], accumulate=False)

backprojection_skin1_map = cv2.calcBackProject([train_hsv], [0], skin1_hist, [0, 180], scale=1)
backprojection_skin2_map = cv2.calcBackProject([train_hsv], [0], skin2_hist, [0, 180], scale=1)

neighborhood_diameter = 3
sigmaColor = 0.25
sigmaSpace = 100

threshold = 0.000001
mask1 = cv2.bilateralFilter(backprojection_skin1_map, neighborhood_diameter,
sigmaColor, sigmaSpace)
mask2 = cv2.bilateralFilter(backprojection_skin2_map, neighborhood_diameter,
sigmaColor, sigmaSpace)
mask1 = mask1 > threshold
mask2 = mask2 > threshold
# mask1 += mask2
masked_image = np.copy(images_in_rgb_format(PATHIMAGE))
masked_image[mask1 == 0] = [0, 255, 0]
plt.imshow(masked_image)



```
